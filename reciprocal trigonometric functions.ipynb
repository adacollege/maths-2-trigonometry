```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
import numpy as np
plt.xkcd();
```

```python
fig, ax = plt.subplots(1,3)
fig.set_figwidth(25)
fig.set_figheight(8)
fig.suptitle("Write expressions for each of the coloured lengths", fontsize=24)

# Fig 1
ax[0].set_xlim(-1,2)
ax[0].set_ylim(-1,2)
ax[0].set_aspect('equal')
ax[0].axis('off')
# Axes and unit circle
ax[0].axhline(0,alpha=0.2)
ax[0].axvline(0,alpha=0.2)
ax[0].add_patch(Arc(xy=(0,0),width=2,height=2,angle=0,theta1=0,theta2=90))

# Define point on circle
a = np.pi/3
s = np.sin(a)
c = np.cos(a)

# Colour relevant lengths
ax[0].plot([0,c],[0,s],'b')
ax[0].plot([c,1/c],[s,0],'y')
ax[0].plot([c,c],[0,s],'r')
ax[0].plot([0,c],[0,0],'g')
ax[0].plot([c,1/c],[0,0],'m')
# Mark angle and hypotenuse
ax[0].annotate("$\\theta$",xy=(0,0),textcoords=('offset points'),xytext=(10,3),fontsize=18)
ax[0].annotate("$1$",xy=(c/2,s/2),textcoords=('offset points'),xytext=(-10,1),fontsize=18)

# Fig 2
ax[1].set_aspect('equal')
ax[1].set_xlim(-1,2)
ax[1].set_ylim(-1,2)
ax[1].axis('off')

# Pull out first triangle
ax[1].plot([0,c],[0,s],'b')
ax[1].plot([c,c],[0,s],'r')
ax[1].plot([0,c],[0,0],'g')
# Mark angle and hypotenuse
ax[1].annotate("$\\theta$",xy=(0,0),textcoords=('offset points'),xytext=(10,3),fontsize=18)
ax[1].annotate("$1$",xy=(c/2,s/2),textcoords=('offset points'),xytext=(-10,1),fontsize=18)

# Fig 3
ax[2].set_aspect('equal')
ax[2].set_xlim(-1,2)
ax[2].set_ylim(-1,2)
ax[2].axis('off')

# Pull out second triangle
ax[2].plot([-s,0],[0,0],'r')
ax[2].plot([0,0],[0,1/c-c],'m')
ax[2].plot([-s,0],[0,1/c-c],'y');
```
